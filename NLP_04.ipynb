```python
import pandas as pd
import numpy as np
import re

```

```python
data_out = pd.read_csv('/content/twitter4000.csv')
```

```python
pattern = r'\w+@\w+.\w+'
```

```python
data_out['email'] = data_out['twitts'].apply(lambda x: re.findall(pattern, x))
data_out['email'].sample(5)
```

```python
import re
pattern = r'\s'
data_out['twitts'] = data_out['twitts'].apply(lambda x: re.sub(pattern, '', x))
```

```python
pattern = r'\b[A-Za-z._+-]+@[A-Za-z.][A-Z|a-z]{2,}\b'
data_out['email'] = data_out['twitts'].apply(lambda x: re.findall(pattern, x))
data_out['email'].sample(5)
```

```python
import re
data = data_out
pattern = r'\b[A-Za-z._+-]+@[A-Za-z.][A-Z|a-z]{2,}\b'
data['email'] = data['twitts'].apply(lambda x: re.findall(pattern, x))
data.head()
```

```python
import re
pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
data['email'] = data['twitts'].apply(lambda x: ",".join(re.findall(pattern, x)))
data['email_count'] = data['twitts'].apply(lambda x: len(re.findall(pattern, x)))
data[['twitts', 'email', 'email_count']].head()
```

```python
url_pattern = r'http\S+|www\S+'
data['urls'] = data['twitts'].apply(lambda x: ",".join(re.findall(url_pattern, str(x))))
data['url_count'] = data['urls'].apply( lambda x: x.count(',') + 1 if x != '' else 0)
data
```
