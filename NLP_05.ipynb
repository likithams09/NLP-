```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
para = """Bluetooth is a wireless technology used to connect devices over short distances without cables.
It allows easy data transfer, audio streaming, and device control using radio waves.
Bluetooth is energy-efficient and widely used in phones, headphones, and smart devices."""
```

```python
sentence = nltk.sent_tokenize(para)
sentence
```

**stemming**

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
for i in range(len(sentence)):
  words = nltk.word_tokenize(sentence[i])
  words = [stemmer.stem(word) for word in words]
  sentence[i] = ' '.join(words)
```

**Lemmatization**

```python
from nltk.stem import WordNetLemmatizer
```

```python
para1 = """Bluetooth is a wireless technology used to connect devices over short distances without cables.
It allows easy data transfer, audio streaming, and device control using radio waves.
Bluetooth is energy-efficient and widely used in phones, headphones, and smart devices."""
```

```python
sentence1 = nltk.sent_tokenize(para1)
```

```python
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
for i in range(len(sentence1)):
  words = nltk.word_tokenize(sentence1[i])
  newwords = [lemmatizer.lemmatize(word) for word in words]
  sentence1[i] = ' '.join(newwords)
sentence1
```
