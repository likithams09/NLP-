```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
input = "Hi I am likii . how are you"
output1 = word_tokenize(input)
print(output1)
```

```python
output2 = sent_tokenize(input)
print(output2)
```
